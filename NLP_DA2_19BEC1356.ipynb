# Conert to Lowercase

```python
text=text.lower()
text
```

# sentence Tokenization

```python
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')
```

```python
sent=nltk.sent_tokenize(text)
sent
```

# Word Tokenization

```python
word=[]
for i in sent:
  word.append(nltk.word_tokenize(i))
word
```

```python
p=[]
for i in word:
  for j in i:
    p.append(j)
p
```

# Stopword Removal

```python
nltk.download('stopwords')
```

```python
tokens=p
from nltk.corpus import stopwords
lang_stopwords = stopwords.words("english")
remove_stopwords = [token for token in tokens if token not in lang_stopwords]
remove_stopwords
```

# Remove Punctuators

```python
import string
token=remove_stopwords
remove_punctuations = [token for token in tokens if not token in string.punctuation]
remove_punctuations
```

# Pos tagging

```python
pos=nltk.pos_tag(remove_punctuations)
pos
```

# Lemmatization

```python
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
from nltk.stem import WordNetLemmatizer 
lemmatizer = WordNetLemmatizer()
tokens=remove_punctuations
Lemma = [lemmatizer.lemmatize(word) for word in tokens]
Lemma
```

# Stemming

```python
from nltk.stem.api import StemmerI
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
token=Lemma
stemmer = PorterStemmer()
stemming_tokens = [stemmer.stem(token) for token in tokens]
stemming_tokens
```
